```python
# Define column names
columns = ['yr', 'mo', 'dy', 'hr', 'mn', 'wdir', 'wspd', 'gst', 'wvht', 'dpd', 'apd', 'mwd', 'pres', 'atmp', 'wtmp', 'dewp', 'vis', 'tide']
```

```python
dtypes = {
    'yr': int, 'mo': int, 'dy': int, 'hr': int, 'mn': int, 
    'degT': float, 'm/s': float, 'm/s': float, 'm': float, 
    'sec': float, 'sec': float, 'degT': float, 'hPa': float, 
    'degC': float, 'degC': float, 'degC': float, 'mi': float, 'ft': float
}
```

```python
# get the data in a dataframe
waves_df = pd.read_csv("2023_waves_data.txt", delim_whitespace=True, names=columns, dtype=dtypes)
waves_df.describe()
```

```python
y = waves_df['wvht']
```

```python
wave_features = ['hr', 'mn', 'wdir', 'wspd']
```

```python
X = waves_df[wave_features]
X.describe()
```

```python
# PREDICTION LOGIC:

# list to hold next data points
next_five_waves = []

# print the next 5 waves
for i in range(5):    

    # set training variables
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 42)

    # define the model
    waves_model = RandomForestRegressor(random_state=42-i)

    # fit the model
    waves_model.fit(train_X, train_y)

    # predict the wave height
    waves_prediction = waves_model.predict(val_X)

    #add the wave height to the list
    next_five_waves.append(round(waves_prediction[-1], 3))
```

```python
# print the next 5 waves
print("Next 5 Waves: ")
for i in range(len(next_five_waves)):
    print(f"Wave {i}: {next_five_waves[i]} Meters or {round((next_five_waves[i] * 3.2808), 3)} Feet")
```
